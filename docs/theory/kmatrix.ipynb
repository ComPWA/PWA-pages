## Definition of the $K$-matrix

$2\times2$ $K$-matrix:

```python
k11, k12, k21, k22 = sy.symbols("k11 k12 k21 k22")
K = sy.Matrix([[k11, k21], [k12, k22]])
K
```

$S$-matrix:

```python
S = (sy.eye(2) + sy.I * K) * (sy.eye(2) - sy.I * K).inv()
S
```

$T$-matrix:

```python
T = K * (sy.eye(2) - sy.I * K).inv()
T
```
